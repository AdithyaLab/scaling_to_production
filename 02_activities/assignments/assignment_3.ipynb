# Load the data

Assuming that the files `adult.data` and `adult.test` are in `../data/adult/`, then you can use the code below to load them.

```python
# import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import OneHotEncoder , StandardScaler , RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import train_test_split , GridSearchCV
```

```python
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = None, names = columns, skiprows=1))
fires_dt
```

# Get X and Y

Create the features data frame and target data.

```python
X = fires_dt.drop(columns='area')
Y = fires_dt[['area']]

print('features df: ', X.shape)
print('target df: ', Y.shape)
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
# preproc1
preproc1 = ColumnTransformer([
    ('num_transformer', StandardScaler(),['ffmc','dmc','dc','isi','temp','rh','wind','rain']),
    ('col_transformer', OneHotEncoder(handle_unknown='infrequent_if_exist'),['month'])
])
preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
# preproc2
preproc2 = ColumnTransformer([
    ('num_transformer', RobustScaler(),['ffmc','dmc','dc','isi','temp','rh','wind','rain']),
    ('col_transformer', OneHotEncoder(handle_unknown='infrequent_if_exist'),['month'])
])
preproc2
```

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.

```python
# Pipeline A = preproc1 + baseline

pipe_p1_base = Pipeline([
    ('preproc1',preproc1),
    ('nearest_neighbour' , KNeighborsRegressor())
])
pipe_p1_base
```

```python
pipe_p1_base.get_params()
```

```python
# Pipeline B = preproc2 + baseline
pipe_p2_base = Pipeline([
    ('preproc2', preproc2),
    ('nearest_neighbour' , KNeighborsRegressor())
])
pipe_p2_base
```

```python
pipe_p2_base.get_params()
```

```python
# Pipeline C = preproc1 + advanced model
pipe_p1_adv = Pipeline([
    ('preproc1',preproc1),
    ('adaboost_reg', AdaBoostRegressor())
])
pipe_p1_adv
```

```python
pipe_p1_adv.get_params()
```

```python
# Pipeline D = preproc2 + advanced model
pipe_p2_adv = Pipeline([
    ('preproc2',preproc2),
    ('adaboost_reg', AdaBoostRegressor())
])
pipe_p2_adv
```

```python
pipe_p2_adv.get_params()
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
x_train, x_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
scoring = ['neg_root_mean_squared_error'] 

base_params = {
    'nearest_neighbour__n_neighbors': [ 5 , 6],
    'nearest_neighbour__p' : [2],
    'nearest_neighbour__weights': [ 'uniform' , 'distance' ]
}

advanced_params = {
    'adaboost_reg__estimator': ['None'],
    'adaboost_reg__loss' : ['linear' , 'square'],
    'adaboost_reg__n_estimators': [50, 100]
}
```

```python

grid_cv = GridSearchCV(
    estimator=pipe_p1_base, 
    param_grid=base_params, 
    scoring = scoring, 
    cv = 5,
    refit = "neg_log_loss")
grid_cv.fit(x_train, y_train)
```
